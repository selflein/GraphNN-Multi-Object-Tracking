```python
from src.tracker.data_track import MOT16
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
dataset = MOT16('./data/MOT16', 'train')
```

```python
item = dataset[1][1]
item
```

```python
fig, ax = plt.subplots()

plt.imshow(item['img'].numpy().transpose(1, 2, 0))

gt = item['gt']
for gt_id, box in gt.items():
    rect = plt.Rectangle(
      (box[0], box[1]),
      box[2] - box[0],
      box[3] - box[1],
      fill=False,
      linewidth=1.0)
    ax.add_patch(rect)
```

```python
box_img = []
for sequence in dataset:
    for item in tqdm(sequence):
        gt = item['gt']
        img = item['img']
        for gt_id, box in gt.items():
            h, w, c = img.shape
            box = box.astype(np.int).clip(0, None)
            crop = img[:, box[1]:box[3], box[0]:box[2]]
            box_img.append((box, crop))
    break
```
